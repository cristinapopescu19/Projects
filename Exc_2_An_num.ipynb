EX#2 (a) Creati in Python functiile Secantaf si PozitieFalsaf care au ca date de intrare

(i) functia f care defineste ecuatia neliniara,
(ii) aproximarile initiale x0 si x1, si
(iii) toleranta admisa TOL, si data de iesire $x_{aprox}$, generata de metoda secantei,
respectiv de metoda pozitiei false (regula falsi), folosind structura repetitiva *while*
si criteriul de oprire $|f(x_n)|$ â‰¤ TOL.

```python
import numpy as np
import matplotlib.pyplot as plt
f = lambda x: x + np.exp(-x**2)*np.cos(x)
def Secantaf(f, x0, x1, TOL, N):
    x = np.zeros(N+1) # contine valorile aproximarilor radacinii in timpul iteratiei
    x[0] = x0
    x[1] = x1
    err = np.zeros(N) #erorile relative
    err[0] = np.abs(x[0] - x[1]) / np.abs(x[0]) #prima eroare relativa
    n = 1
    print(f"n = {n} | x{n} = {x[n]} | err = {err[n-1]}") #prima iteratie

    while n < N and np.abs(f(x[n])) > TOL: #while ruleaza cat nr iteratiilor e mai mic decat N si cat valoarea in x[n] e mai mare decat toleranta
        n += 1
        x[n] = x[n-1] - f(x[n-1]) * (x[n-1] - x[n-2]) / (f(x[n-1]) - f(x[n-2]))
        err[n-1] = np.abs(x[n-1] - x[n]) / np.abs(x[n-1])
        print(f"n = {n} | x{n} = {x[n]} | err = {err[n-1]}")

    return x, err #returnam aproximarea radacinii si erorile relative

def PozitieFalsaf(f, x0, x1, TOL, N):
    x = np.zeros(N+1)
    n = 1
    x[0] = x0
    x[1] = x1
    err = np.zeros(N)
    err[n-1] = np.abs(x[n-1]-x[n])/np.abs(x[n-1])
    print(f"n = {n} | x{n} = {x[n]} | err = {err[n-1]}")

    while n < N and np.abs(f(x[n])) > TOL:
        n += 1
        x[n] = x[n-1] - f(x[n-1]) * (x[n-1] - x[n-2]) / (f(x[n-1]) - f(x[n-2]))
        err[n-1] = np.abs(x[n-1] - x[n]) / np.abs(x[n-1])
        print(f"n = {n} | x{n} = {x[n]} | err = {err[n-1]}")

    return x, err

TOL = 10^(-5)
N = 10
x_secant, err_secant = Secantaf(f, -1, 1, TOL, N)
print('\n')
x_fals, err_falsa = PozitieFalsaf(f, -1, 1, TOL, N)

# Reprezentam grafic eroarea relativa ca functie de n pentru metoda secantei
n_secant = np.arange(1, len(err_secant) + 1)
plt.figure(figsize=(5, 5))
plt.plot(n_secant, err_secant, marker='o')
plt.xlabel('n')
plt.ylabel('Eroarea relativa')
plt.title('Eroarea relativa pentru metoda secantei')
plt.grid(True)


# Reprezentam grafic eroarea relativa ca functie de n pentru metoda pozitiei false
n_false = np.arange(1, len(err_falsa) + 1)
plt.figure(figsize=(5, 5))
plt.plot(n_false, err_falsa, marker='o')
plt.xlabel('n')
plt.ylabel('Eroarea relativa')
plt.title('Eroarea relativa pentru metoda pozitiei false')
plt.grid(True)

plt.show()
```
